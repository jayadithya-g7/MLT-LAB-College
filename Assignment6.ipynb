**<br> Simple Linear Regression <br>**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Load datasets
wine_data = pd.read_csv('/content/winequality-red[1].csv', sep=';')
insurance_data = pd.read_csv('/content/insurance[1].csv')
```

**<br> Wine dataset analysis <br>**

```python
# Wine dataset analysis
print("\nWine Dataset Analysis")
print(wine_data.info())

# Check for null values
wine_nulls = wine_data.isnull().sum()
print("\nNull values in wine dataset:\n", wine_nulls)

# Simple Linear Regression for wine dataset
X_wine = wine_data.drop('quality', axis=1)
y_wine = wine_data['quality']

regression_models = {}
for feature in X_wine.columns:
    model = LinearRegression()
    X_feature = X_wine[[feature]]
    model.fit(X_feature, y_wine)
    regression_models[feature] = (model.intercept_, model.coef_[0])

print("\nRegression equations for wine dataset:")
for feature, (intercept, slope) in regression_models.items():
    print(f"quality = {intercept:.3f} + {slope:.3f}*{feature}")
```

**<br> Insurance dataset Analysis<br>**

```python

# Insurance dataset analysis
print("\nInsurance Dataset Analysis")
print(insurance_data.info())

# Check for null values
insurance_nulls = insurance_data.isnull().sum()
print("\nNull values in insurance dataset:\n", insurance_nulls)

# Encode categorical data
categorical_features = ['sex', 'smoker', 'region']
encoder = OrdinalEncoder()
insurance_data[categorical_features] = encoder.fit_transform(insurance_data[categorical_features])

# Simple Linear Regression for insurance dataset
features = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']
y_insurance = insurance_data['charges']

regression_models_insurance = {}
for feature in features:
    model = LinearRegression()
    X_feature = insurance_data[[feature]]
    model.fit(X_feature, y_insurance)
    regression_models_insurance[feature] = (model.intercept_, model.coef_[0])

print("\nRegression equations for insurance dataset:")
for feature, (intercept, slope) in regression_models_insurance.items():
    print(f"charges = {intercept:.3f} + {slope:.3f}*{feature}")
```
